```python
import pytrends
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
btcinterest = pytrends.get_historical_interest(['BTC'], year_start=2022, month_start=3, day_start=1, hour_start=0, \
                                 year_end=2022, month_end=9, day_end=25, hour_end=0, cat=0, geo='', gprop='', sleep=0)
btcinterest.plot()
```
