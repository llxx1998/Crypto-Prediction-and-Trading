# ETL for Price and Vol Data
## Get top 50 coins
Note: we find later that lunc only has 37 rows of data, much lower than other coins. So 51th coin is used to replace it

```python
import re
import pandas as pd

with open("cryptohtml.txt") as f:
    text = f.readlines()
text = "\n".join(text)

result = re.findall(r"(?<=data-coin-symbol=\").*?(?=\")", text)
# remove duplicates
pos = 1
while pos < len(result):
    if result[pos] == result[pos-1]:
        del result[pos]
    else:
        pos += 1

print(result, len(result))
temp = result[50]
result.remove("lunc")
pd.Series(result[:50]).to_csv("./top_50_coin_ticker.csv")
```

## Yahoo query

```python
try:
    from yahooquery import Screener, Ticker
except:
    !pip install yahooquery
    from yahooquery import Screener, Ticker

# get ticker
s = Screener()
data = s.get_screeners('all_cryptocurrencies_us', count=250)
dicts = data['all_cryptocurrencies_us']['quotes']
symbols = [d['symbol'] for d in dicts]
print(len(symbols))
```

```python
result = [x.upper() + "-USD" for x in result[:50]]
ticker_issue = [x for x in result if x not in symbols]
print(result, '\n', ticker_issue)
```

```python
yh_conn = Ticker(result)
res = yh_conn.history(period='730D', interval='1h')
yh_conn
```

```python
res # all 50 are get.
```

```python
length = [len(val) for key, val in res.items()]
min(length), length
```

```python
table = None
for crypto in res.index.get_level_values(0).unique().tolist():
    temp = res.loc[crypto, ]
    temp.columns = [crypto.split("-")[0] + "-" + x for x in temp.columns.tolist()]
    temp.index = temp.index.tz_localize("UTC")
    if table is None:
        table = temp
    else:
        table = pd.merge(table, temp, how='outer', left_index=True, right_index=True)
table.shape
```

## Check and preprocess data
### Data count issue

```python
# consecutive check for date
import datetime
series = table.index.tolist()
cnt = 0
for i in range(1, len(series)):
    if series[i] != series[i-1] + datetime.timedelta(hours=1):
        print(series[i])
        cnt += 1
cnt
```

```python
table['exist'] = 1
ts_complete = pd.DataFrame(index=pd.date_range(start="2020-09-26 00:00:00", end="2022-09-25 04:00:00", freq='1H', tz="UTC"))
table=pd.merge(ts_complete, table, how='left', left_index=True, right_index=True)
for i in range(table.shape[0]):
    if table['exist'].iloc[i] != 1:
        table.iloc[i, :] = table.iloc[i-1, :].to_numpy()
table.drop(columns=['exist'], inplace=True)
table.shape
```

```python
table.to_csv("./price_vol.csv")
```
