```python
import pandas as pd
import numpy as np
```

```python
def ColSelector(df: pd.DataFrame, series=None, toker_list=None):
    cols = df.columns
    res = []
    if series and toker_list:
        res = [x for x in cols if x.split('-')[0] in toker_list and x.split('-')[1] in series]
    elif series:
        res = [x for x in cols if x.split('-')[1] in series]
    elif toker_list:
        res = [x for x in cols if x.split('-')[0] in toker_list]
    return res    
```

```python
df = pd.read_csv('macro.csv')
df.head()
```

```python
df['PeriodDate'] = pd.to_datetime(df['PeriodDate'])
```

```python
df2 = df.pivot_table(values=["Series_Value"], index=["PeriodDate"], columns=["EcoSeriesID"])
df2.columns = df2.columns.droplevel()
hashmap = {1717: "CPI", 
           940: "GDP"}
df2.rename(columns=hashmap, inplace=True)
df2['GDP'] = df2['GDP'].pct_change()
df2
```

```python
data = pd.read_csv("price_vol.csv", index_col=0)
data.index=pd.to_datetime(data)
```

```python
# replace negative value
for colname, val in data.iteritems():
    val[val <= 0] = np.nan
```

```python
data.head()
```

```python
cols = ColSelector(data, series=['low', 'high', 'volume', 'close'])
data[cols] = data[cols].pct_change()
data
```

```python
# fillna
## step 1: missing in the middle
data.fillna(method="ffill", inplace=True)
## step 2: filling in the front
data.fillna(method="bfill", inplace=True)
data
```

```python
data.to_csv('price_vol_filledNa.csv', index=True)
```
